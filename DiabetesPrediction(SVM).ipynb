**Importing the Dependencies**

```python
import pandas as pd 
import numpy as np  # -> used for creating data frame  
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score 
#standizer function                 
from sklearn.preprocessing import StandardScaler   # -> standardize the data to a common range 
from sklearn import svm
```

**Data Collection and Analysis 
PIMA Data Sets**

```python
# -> This data is for only females
diabetes_data = pd.read_csv("diabetes.csv")
# -> use read_csv? to know more about the function 
#pd.read_csv?
```

```python
# -> printing the first 5 rows of the data sets 
diabetes_data.head()
```

```python
# -> Number of rows and columns in tha data set 
diabetes_data.shape
```

```python
# -> getting the statistiacl measures of the data
diabetes_data.describe()
```

```python
diabetes_data['Outcome'].value_counts
```

**0 -> Non Diabetic**
**1 -> Diabetic**

```python
# -> Calculating the mean value for all the data set
diabetes_data.groupby('Outcome').mean()
```

```python
# -> Seperating the data into lables 
x = diabetes_data.drop(columns ='Outcome' , axis = 1) 
y = diabetes_data['Outcome']
```

```python
print(x)
print(y)
```

**Data Standrization**

```python
scaler =StandardScaler()
```

```python
scaler.fit(x)
```

```python
# -> We make the standardization steps to make all values at the same Range
standarized_data = scaler.transform(x)
# -> you can do standarized_data = scaler.fit_transform(x) to do those 2 steps into one step
```

```python
print(standarized_data)
# -> all the data at the range 0 and 1
```

```python
# -> X for the data
x = standarized_data
# -> Y for the model
y = diabetes_data['Outcome']
```

```python
print(x)
print(y)
```

**Splitting The Data**

```python
# -> stratify= y to make sure the splitting label data is from y each time we run the code (splitting the data in the same proportion )
# -> random_state = 2 to make sure that they will split in the same way each time we run the code
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.2 , stratify= y, random_state = 2) 
```

```python
print(x.shape , x_train.shape , x_test.shape)
```

**Training The Model** 

```python
classifier = svm.SVC(kernel='linear')
```

```python
# -> Training The Support Vector Machine Classifier
classifier.fit(x_train ,y_train)
```

**Model Evaluation**

**Accuracy Score**

```python
# -> The accuracy of the training data will be more but the test data will be less

x_train_prediction = classifier.predict(x_train)
```

```python
# -> compare the accuracy values of X_train with the actual value y_train
# -> the accuracy on the training data will be higher, but the accuracy on the test data will be lower. This is a sign of overfitting.


training_data_accuracy = accuracy_score(x_train_prediction ,y_train)
print("Accuracy score of the training data", training_data_accuracy)
```

```python
x_test_prediction = classifier.predict(x_test)
```

```python
test_data_accuracy = accuracy_score(x_test_prediction ,y_test)
print("Accuracy score of the test data", test_data_accuracy)
```

**Making a Predictive System**

```python
input_data = (13,126,90,0,0,43.4,0.583,42)

input_data_as_numby_array = np.asarray(input_data)

input_data_reshape = input_data_as_numby_array.reshape(1,-1)

# -> Because we have transformed the data before we have to do the same thing with that input 
# -> Standardize the data  
std_data = scaler.transform(input_data_reshape)
print(std_data)

prediction = classifier.predict(std_data)
print(prediction)


# prediction[0] ->= Because prediction is a list [0] first element in the list

if (prediction[0]== 0):
    print("The person is not diabetic")
else :
    print("The person is diabetic") 
```
